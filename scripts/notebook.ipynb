```python
api_url = "http://localhost:5000/get-file"
payload = {
    "file":"bdo.csv"
}
```

```python
response = requests.post(api_url, json=payload)
```

```python
response.text
```

```python
data = StringIO(response.text)
df = pd.read_csv(data)
df
```

```python
df
```

```python
buffer = io.StringIO()
df.to_csv(buffer, index=False)
buffer.seek(0)

response = requests.post("http://localhost:5000/create-silver", 
                         data=buffer.getvalue(),
                         headers={"Content-Type": "text/csv"})
```
